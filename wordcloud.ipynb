```python
def displayWordCloud(data = None, backgroundcolor = 'black', width=800, height=600 ): 
    wordcloud = WordCloud( font_path = 'C:\\Users\\admin\\AppData\\Local\\Microsoft\\Windows\\Fonts\\NanumGothicEcoR.ttf', stopwords= stopwords_kr, background_color = backgroundcolor, width = width, height = height).generate(data) 
    print(wordcloud.words_) 
    plt.figure(figsize = (15, 10)) 
    plt.imshow(wordcloud) 
    plt.axis("off") 
    plt.show()

```
